```python
#necesary library for this analisys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
#link of dataset: https://www.kaggle.com/datasets/lehaknarnauli/spotify-datasets?resource=download
```

```python
#upload data tracks
tracks = pd.read_csv(r"tracks.csv")
#show data
tracks.head(8)
```

```python
#df info
tracks.info()
```

```python
#show data ascending according to column popularity
df_tracks = tracks.sort_values("popularity", ascending=True)
df_tracks.head(5)
```

```python
#most popular sound tracks with filter query
most_popular = tracks.query("popularity>90", inplace=False).sort_values("popularity", ascending=False)
#show the first 10 data
most_popular.head(10)
```

```python
import random
#Function to generate a random day and month and keep the original year
def completar_fecha(fecha):
    partes = fecha.split('-')
    if len(partes) == 1:
        #Contains only the year, a random day and month are added
        dia = random.randint(1, 28)  #Assuming all months have 28 days
        mes = random.randint(1, 12)
        return f'{fecha}-{mes:02d}-{dia:02d}'
    elif len(partes) == 2:
        #Contains year and month, a random day is added
        dia = random.randint(1, 28)
        return f'{fecha}-{dia:02d}'
    else:
        #Date already complete, returned as is
        return fecha

#Apply the function to the necessary cells
tracks['release_date'] = tracks['release_date'].apply(completar_fecha)

tracks.head()
```

```python
#select column release data like index of dataframe
tracks.set_index("release_date", inplace=True)
#convert index to date time
tracks.index = pd.to_datetime(tracks.index)
#show changes
tracks.head()
```

```python
#this metod select column artist and show information of the row 97
tracks[["artists"]].iloc[97]
```

```python
#convert data ms to seconds
tracks["duration"] = tracks["duration_ms"].apply(lambda x: round(x/1000))
#drop column
tracks.drop("duration_ms", inplace=True, axis=1)
#show result
tracks.head()
```

```python
#correlation heat map
corr_df = tracks.drop(["key", "mode", "explicit", "artists", "id_artists", "name", "id"], axis=1)
corr_df = corr_df.corr(method="pearson")
plt.figure(figsize=(14,6))
sns.heatmap(corr_df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1, linecolor="Black")
plt.title("Correlation heatmap between variable")
plt.show()
```

```python
#data wit 0.5% sample of original data
sample_df = tracks.sample(int(0.005*len(tracks)))
#show tall of data
print(len(sample_df))
```

```python
#graphic regplot loudness vs energy
plt.figure(figsize=(14,6))
sns.regplot(data=sample_df, y="loudness", x="energy")
plt.title("Loudness vs Energy Correlation")
plt.show()
```

```python
#graphic regplot danceability vs popularity
plt.figure(figsize=(14,6))
sns.regplot(data=sample_df, y="energy", x="acousticness")
plt.title("acousticness vs energy Correlation")
plt.show()
```

```python
#index to column
tracks["dates"]=tracks.index.get_level_values("release_date")
tracks.dates=pd.to_datetime(tracks.dates)
years=tracks.dates.dt.year
```

```python
#displot graphic
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist")
plt.title("Number of song per year")
plt.show()
```

```python
#barplot graphic
total_dr = tracks.duration
plt.figure(figsize=(18,7))
plt.subplot(1,1,1)
sns.barplot(x=years, y=total_dr)
plt.title("Year vs duration")
plt.xticks(rotation=90)
plt.show()
```

```python
#lineplot graphic
total_dr = tracks.duration
sns.set_style(style="whitegrid")
plt.figure(figsize=(14,6))

sns.lineplot(x=years, y=total_dr)
plt.title("Year vs duration")
plt.xticks(rotation=90)
plt.show()
```

```python
#graphic barplot
sns.set_style(style="darkgrid")
famous = tracks.sort_values("popularity", ascending=False).head(10)
sns.barplot(x="popularity", y="name", data=famous)
plt.title("Top sound most popular")
plt.show()
```

```python
artist = pd.read_csv(r"artists.csv")
artist.head()
```

```python
#artist most popular
#graphic barplot
sns.set_style(style="darkgrid")
famous = artist.sort_values("popularity", ascending=False).head(10)
sns.barplot(x="popularity", y="name", data=famous)
plt.title("Top artists most popular")
plt.show()
```

```python
#artist most follower
top_followed = artist.nlargest(5, "followers")
plt.bar(top_followed["name"], top_followed["followers"])
plt.title("Most followed artist")
for i, v in enumerate(top_followed['followers']):
    plt.text(i, v + 5, str(v), ha='center', va='bottom')
plt.show()
```

```python
#genres most popular
#graphic barplot
sns.set_style(style="darkgrid")
famous = artist.sort_values("popularity", ascending=False).head(10)
sns.barplot(x="popularity", y="genres", data=famous)
plt.title("Top genres most popular")
plt.show()
```
